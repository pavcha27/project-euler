```python
import pandas as pd
import numpy as np
```

```python
normal_months = {
    1: 31, 2: 28, 3: 31,
    4: 30, 5: 31, 6: 30,
    7: 31, 8: 31, 9: 30,
    10: 31, 11: 30, 12: 31
}

leap_months = {
    1: 31, 2: 29, 3: 31,
    4: 30, 5: 31, 6: 30,
    7: 31, 8: 31, 9: 30,
    10: 31, 11: 30, 12: 31
}
```

```python
def add_week(day, months_dict):
    next_year = day[0]
    next_month = day[1]
    next_day = day[2] + 7
    if next_day > months_dict[next_month]:
        next_day-= months_dict[next_month]
        if next_month == 12:
            next_year += 1
            next_month = 1
        else:
            next_month += 1
    
    return [next_year, next_month, next_day]
```

```python
leap_years = []
for i in range(1901, 2001):
    if (i % 400 == 0) or (i % 4 == 0 and i % 100 > 0):
        leap_years.append(i)
```

```python
sunday = [1901, 1, 6]
sundays = 0
while sunday[0] < 2001:
    print(sunday)
    
    if sunday[2] == 1:
        sundays += 1
    
    sunday = add_week(sunday, leap_months if sunday[0] in leap_years else normal_months)
```
