```python
import pandas as pd
import numpy as np

from collections import Counter
```

```python
def sieve(n):
    arr = [0,0,1] + [1,0]*(n//2 + 1)
    i = 3
    while i*i <= n:
        if arr[i]:  
            arr[i*i::2*i] = [0]*len(arr[i*i::2*i])
        i += 2

    ret = []
    for (i, p) in enumerate(arr):
        if p:
            ret.append(i)

    return arr, ret
```

```python
def is_permutation(n1, n2):
    return Counter(str(n1)) == Counter(str(n2))
```

$\phi(p) = p - 1$ whenever $p$ is prime, and $p-1$ cannot be a permutation of $p$. So we instead check all products of $2$ primes next. Given $p_i, p_j$ prime, if $n = p_i p_j$, then we get
$$
\begin{align*}
\phi(n) &= n\left(1 - \frac{1}{p_i}\right)\left(1 - \frac{1}{p_j}\right) = (p_i - 1) (p_j - 1), \\ \\
\frac{n}{\phi(n)} &= \frac{n}{n\left(1 - \frac{1}{p_i}\right)\left(1 - \frac{1}{p_j}\right)} = \frac{p_i p_j}{(p_i - 1)(p_j - 1)} \\ \\
& \phi(n) \text{ is permutation of } n
\end{align*} 
$$

```python
pbs, ps = sieve(4*10**3)
ratio = 3
best_pi, best_pj = -1, -1
for i in range(len(ps))[::-1]:
    pi = ps[i]
    
    for j in range(i+1, len(ps)):
        pj = ps[j]
        n = pi*pj
        if n > 10**7:
            break

        if is_permutation(n - pi - pj + 1, n):
            # br = True
            print(pi, pj, n)
            print((pi - 1)*(pj - 1))
            print(pi*pj / (n - pi - pj + 1))
            print()

            if pi*pj / (n - pi - pj + 1) < ratio:
                ratio = pi*pj / (n - pi - pj + 1)
                best_pi = pi
                best_pj = pj
            
            break

print((best_pi, best_pj, best_pi*best_pj))
```
