```python
import pandas as pd
import numpy as np
import random
```

```python
def sieve(n):
    arr = np.array([0,0,1] + [1,0]*(n//2 + 1))
    i = 3
    while i*i <= n:
        if arr[i]:  
            arr[i*i::2*i] = [0]*len(arr[i*i::2*i])
        i += 2

    return arr
```

```python
def miller_rabin(n, k = 10):
    if n <= 3:
        return n == 2 or n == 3

    if n % 2 == 0:
        return False

    r, s = 0, n - 1
    while s % 2 == 0:
        r += 1
        s //= 2
    for _ in range(k):
        a = random.randrange(2, n - 1)
        x = pow(a, s, n)
        if x == 1 or x == n - 1:
            continue
        for _ in range(r - 1):
            x = pow(x, 2, n)
            if x == n - 1:
                break
        else:
            return False
    return True
```

```python
def num_spiral(side_length):
    if side_length % 2 == 0:
        raise ValueError("side lengths must be odd!")
    
    grid = np.zeros((side_length, side_length,))

    directions = [(0,1), (-1,0), (0,-1), (1,0)]
    direction = 0

    curr_row = side_length//2
    curr_col = side_length//2
    grid[curr_row, curr_col] = 1
    i = 2

    while curr_row < side_length-1 or curr_col < side_length-1:
        curr_row += directions[direction][0]
        curr_col += directions[direction][1]
        grid[curr_row, curr_col] = i

        # if there is something in the next direction, continue straight
        new_direction = (direction + 1) % 4
        if not grid[curr_row + directions[new_direction][0], curr_col + directions[new_direction][1]]:
            direction = new_direction
        
        i += 1

    return grid
```

```python
pbs = sieve(10**6)
```

```python
sl = 1
new_set = {1}
prime_cnt = 0
diag_cnt = 1

while True:
    sl += 2
    slsq = sl*sl
    slm1 = sl - 1
    for i in range(4):
        if slsq < len(pbs):
            prime_cnt += pbs[slsq - i*slm1]
        else:
            prime_cnt += 1 if miller_rabin(n = slsq - i*slm1, k=10) else 0
        diag_cnt += 1

    print(sl, ":", prime_cnt/diag_cnt)

    if prime_cnt/diag_cnt < 0.1:
        print(sl)
        break
```
