```python
import pandas as pd
import numpy as np

from itertools import product
from functools import reduce
```

```python
def find_all_divisors(num):
    divs = [1]
    i = 2
    while i*i <= num:
        if not num % i:
            divs.append(i)
        
        i += 1
    
    n = len(divs)
    for j in range(n)[::-1]:
        divs.append(num // divs[j])
    
    ret = []
    for d in divs:
        if d < num:
            ret.append(d)

    return set(ret)
```

```python
abundant_numbers = []
for i in range(12, 28124 - 12):
    if sum(find_all_divisors(i)) > i:
        abundant_numbers.append(i)

nums = [1]*28124
for i in range(len(abundant_numbers)):
    num1 = abundant_numbers[i]
    for j in range(i, len(abundant_numbers)):
        num2 = abundant_numbers[j]
        if num1 + num2 > 28123:
            break
            
        nums[num1 + num2] = 0
```

```python
for i in range(1,len(nums)):
    if nums[i] == 1:
        print(i)
```
